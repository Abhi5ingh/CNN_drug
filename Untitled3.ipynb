```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

model = tf.keras.models.load_model("cnn_tki_classifier.h5")

val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
val_generator = val_datagen.flow_from_directory(
    "dataset",          # same folder you used for training
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="validation",
    shuffle=False
)

# Then compute confusion matrix as before
pred_probs = model.predict(val_generator)
...
```

```python
pred_probs
```

```python
from sklearn.metrics import confusion_matrix, classification_report



# Convert probabilities to class labels (0 or 1) using a threshold of 0.5
pred_labels = (pred_probs > 0.5).astype("int32").ravel()

# The ground-truth labels from the generator
true_labels = val_generator.classes  # For a DirectoryIterator, .classes gives the true labels
```

```python
cm = confusion_matrix(true_labels, pred_labels)
print("Confusion Matrix:")
print(cm)

```

```python
# 5. Classification report
print("Classification Report:")
print(classification_report(true_labels, pred_labels, target_names=["inactive", "active"]))
```
